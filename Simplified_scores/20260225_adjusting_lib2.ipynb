```python
import pandas as pd
```

```python
df = pd.read_csv('GMM_classifications/TSC2.lib1.gmm.tsv', sep='\t')

df.head()

df.columns

```

```python
print(df.columns.tolist())
```

```python
simple_score = pd.read_csv('20260217_TSC2_lib1_scores_QC.csv')
simple_score[["ENSP", "variant"]] = simple_score["variant"].str.split(":", n=1, expand=True)
simple_score = simple_score.drop(columns=["ENSP"])
simple_score
```

```python
simple_score = simple_score.merge(
    df[["hgvs_p", "functional_consequence"]].drop_duplicates(),
    left_on="variant",
    right_on="hgvs_p",
    how="left"
).drop(columns=["hgvs_p"])

simple_score
```

```python
simple_score.to_csv("20260217_TSC2_lib1_scores_QC.csv", index=False)
print("Exported to 20260217_TSC2_lib1_scores_QC.csv")
```
